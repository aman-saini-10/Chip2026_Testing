```python
# import required libs
import numpy as np
import sys
import pyvisa
sys.path.append(r"../Communiaction")
sys.path.append(r"../Utils")
sys.path.append(r"../PowerSupply")
import PS_Utils
import TCP_IP_UTILS
import Utils
import time
```

```python
#connect to ethernet socket 
HOST = '192.168.1.10'  # Remote device IP (server IP)
PORT = 7               # Echo port
client_socket = TCP_IP_UTILS.tcp_connect(HOST,PORT)
#Set IOs to default before turning on powersupply
ret_data = Utils.Set_ChipIO_to_Default(client_socket)
print(ret_data)
time.sleep(5) #Wait for IOs to settle
```

```python
#Add power supply code here 
Device_ID_MAIN_PS = "USB0::0x2A8D::0x1202::MY59003914::0::INSTR" #3 channel power supply
rm = pyvisa.ResourceManager()
resources = rm.list_resources()
print("Available VISA resources:")
for r in resources:
        print(" -", r)
#Channel 1 dedicated to VDDA and VDDQ
#Channel 2 dedicated to VDAC and Analog Power Supplies
#Channel 3 dedicated to Board Chip Power Supplies
CHIP_PS = 1
ANA_PS = 2
BOARD_PS = 3
device_main_ps = PS_Utils.connect_E36313A(Device_ID_MAIN_PS)
```

```python
#Set board power supply to 1 volt with 100mA current limit
PS_Utils.set_voltage_E36313A(device_main_ps,BOARD_PS,1,0.1)
time.sleep(1); #Wait for power supply to settle

#Set chip VDDA and VDDQ supply to 1 volt with 100mA current limit
PS_Utils.set_voltage_E36313A(device_main_ps,CHIP_PS,1,0.1)
time.sleep(1); #Wait for power supply to settle

#Set analog supply to 0 volt with 10mA current limit #SRAM needs to be initialised before for VB
PS_Utils.set_voltage_E36313A(device_main_ps,ANA_PS,0,0.01)
time.sleep(1); #Wait for power supply to settle

#Set IOs to default before turning on powersupply
ret_data = Utils.Set_ChipIO_to_Default(client_socket)
print(ret_data)

```

```python
signal_array = [Utils.CLK_A,Utils.CLK_B]
value_array = [0,0]
ret_data = Utils.Set_Chip_IO(client_socket,signal_array,value_array,1)
print(ret_data)
```
