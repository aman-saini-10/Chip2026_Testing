```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'Times New Roman'   # or 'Times New Roman', 'DejaVu Sans', etc.
mpl.rcParams['font.size'] = 15

# Axis-specific defaults
mpl.rcParams['axes.titlesize'] = 15
mpl.rcParams['axes.labelsize'] = 15
mpl.rcParams['xtick.labelsize'] = 15
mpl.rcParams['ytick.labelsize'] = 15
mpl.rcParams['legend.fontsize'] = 15
```

```python
Total_Area = 646*1112
SRAM_Array = 4*598*202
TDC = 2*130*598
DAC = 4*202*33
Control = 18*107
Rest = Total_Area - (SRAM_Array+TDC+DAC) 
```

```python
import matplotlib.colors as mcolors
def lighten_color(color, amount=0.35):
    """
    amount = 0 → original color
    amount = 1 → white
    """
    r, g, b = mcolors.to_rgb(color)
    return (
        r + (1 - r) * amount,
        g + (1 - g) * amount,
        b + (1 - b) * amount
    )

# Original tab10
base_colors = plt.cm.tab10.colors

# Lightened colors
colors = [lighten_color(c, amount=0.25) for c in base_colors]
```

```python

labels = [
    "SRAM Array + Periphery",
    "TDC + Digital Readout",
    "DAC + Activation SRAM",
    "Other"
]
areas = [
    SRAM_Array,
    TDC,
    DAC,
    Rest
]
# ---------------- Plot ----------------
fig, ax = plt.subplots(figsize=(4, 4))

wedges, _ = ax.pie(
    areas,
    colors=colors[:len(areas)],
    startangle=90,
    wedgeprops={'edgecolor': 'black'}
)

ax.legend(
    wedges,
    labels,
    loc='center left',
    bbox_to_anchor=(1.0, 0.5),
    frameon=False
)
#ax.set_title("Area Breakdown")
plt.savefig(r"D:\ISSCC_26\Graphs\Area_Breakdown.jpg", dpi=800, bbox_inches="tight")
plt.show()
```
