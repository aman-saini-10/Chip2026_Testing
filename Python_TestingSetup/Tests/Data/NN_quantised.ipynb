```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset
from scipy.interpolate import interp1d
from matplotlib.ticker import MultipleLocator
```

```python
df = pd.read_excel("Quant_NN.xlsx")
Layers = [1,2,3,4,5,6,10,11,12,13,17,18,19,20]
```

```python
mnist_cnn = (df.iloc[0:5,1:15]).to_numpy()
cifar10_resnet = (df.iloc[6:11,1:15]).to_numpy()
cifar100_resnet = (df.iloc[12:17,1:15]).to_numpy()
data = mnist_cnn
data1 =  cifar10_resnet 
data2 = cifar100_resnet
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm

# Example data
rows, cols = data.shape
x = np.arange(cols)

# Research-friendly colormap (uniform color family)
cmap = cm.seismic  # Try 'Greys', 'Purples', 'cividis', or 'viridis'
colors = [cmap(0.9 - 0.5 * (i / (rows - 1))) for i in range(rows)]

# 🔹 Set figure size (width, height) in inches
plt.figure(figsize=(10, 4))  # Slightly wider for legend space

# Plot
for i in range(rows):
    plt.bar(x, data[i], width=0.6, color=colors[i],
            edgecolor='black', label=f'{i}')

# Formatting
plt.xticks(x, [f'Layer {i}' for i in range(cols)], fontsize=12, rotation=0)
plt.ylabel('Accuracy Percentage (%)', fontsize=14)
plt.ylim(0, 100)
plt.title('MNIST CNN', fontsize=15, pad=10)
plt.grid(axis='y', linestyle='--', alpha=0.4)

# 🔹 Move legend outside the plot area, aligned to the right
plt.legend(title='Information Deficit',fontsize=10,loc='upper left', bbox_to_anchor=(1.02, 0.5),frameon=True, edgecolor='black')

# 🔹 Add right margin so legend fits cleanly
plt.tight_layout(rect=[0, 0, 0.83, 1])
plt.savefig("D:\ISSCC_26\Graphs\MNIST.jpg",dpi=800, bbox_inches="tight", transparent=True)
plt.show()
```
