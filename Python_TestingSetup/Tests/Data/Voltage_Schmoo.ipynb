```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Voltages and max passing frequency at each voltage
voltages   = np.array([0.5, 0.6, 0.7, 0.8, 0.9, 1.0])
max_freqs  = np.array([0, 10.9, 21.8, 34.54, 36.36, 40.0])  # MHz

# Define frequency sweep grid for plotting (from min to max)
freq_grid = np.linspace(0, 50, 500)  # 0–50 MHz at high resolution

# Build pass/fail matrix
results = np.zeros((len(freq_grid), len(voltages)), dtype=int)
for i, v in enumerate(voltages):
    results[:, i] = (freq_grid <= max_freqs[i]).astype(int)  # 1=pass, 0=fail

# Color map
cmap = ListedColormap(['red', 'green'])

# Plot
plt.figure(figsize=(4, 3))
plt.pcolormesh(voltages, freq_grid, results, cmap=cmap, shading='auto')

plt.xticks(voltages)
plt.yticks([0, 10, 20, 30, 40, 50])

plt.xlabel('$V_{DD}$ (V)',fontsize = 12)
plt.ylabel('Frequency (MHz)', fontsize = 12)
plt.title('Voltage Shmoo',fontsize = 12)

plt.ylim(0, 50)
plt.grid(color='black', linewidth=0.7)

plt.tight_layout()

plt.savefig("D:\ISSCC_26\Graphs\Voltage_shmoo.jpg", dpi=800, bbox_inches="tight")
plt.show()
```
