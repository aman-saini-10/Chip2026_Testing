```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset
from scipy.interpolate import interp1d
from matplotlib.ticker import MultipleLocator
```

```python
df = pd.read_csv(r"d:\Chip2025_Testing\Python_Notebook\Tests\Data\CadenceGraphs\NMOS_DISCHG.csv")
df = df[100:]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'Times New Roman'   # or 'Times New Roman', 'DejaVu Sans', etc.
mpl.rcParams['font.size'] = 12

# Axis-specific defaults
mpl.rcParams['axes.titlesize'] = 12
mpl.rcParams['axes.labelsize'] = 12
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 10
```

```python
#fig, ax1 = plt.subplots(figsize=(4, 2))
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True,figsize=(4, 4))
# First plot (left y-axis)

ax1.plot(df["TIME_NMOS"]*(10**12), df["IN"], color="g",linewidth=1,linestyle='--', label = "IN")
ax1.plot(df["TIME_PMOS"]*(10**12), df["OUT_PMOS"], color="b",linewidth=1,linestyle='--', label = "OUT")
ax1.plot(df["TIME_PMOS"]*(10**12), df["CAP_PMOS"], color="r",linewidth=1.5, label = "CAP")
ax1.plot(df["TIME_PMOS"]*(10**12), df["Node_PMOS"], color="purple",linewidth=1.5, label = "Node")



ax1.set_ylabel("Voltage (V)")
#ax1.set_ylabel("Mean Output across Macro", color="r")
#ax1.tick_params(axis="y", labelcolor="r")
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_locator(MultipleLocator(0.3))

ax1.legend(loc = "center right")

ax2.plot(df["TIME_NMOS"]*(10**12), df["IN"], color="g",linewidth=1,linestyle='--', label = "IN")
ax2.plot(df["TIME_NMOS"]*(10**12), df["OUT_NMOS"], color="b",linewidth=1,linestyle='--', label = "OUT")
ax2.plot(df["TIME_NMOS"]*(10**12), df["CAP_NMOS"], color="r",linewidth=1.5, label = "CAP")
ax2.plot(df["TIME_NMOS"]*(10**12), df["Node_NMOS"], color="purple",linewidth=1.5, label = "Node")


ax2.set_ylim(0, 1)
ax2.set_xlim(150, 450)
ax2.yaxis.set_major_locator(MultipleLocator(0.3))
ax2.set_ylabel("Voltage (V)")
ax2.set_xlabel("Time (ps)")
#plt.tight_layout()
ax2.legend(loc = "center right")
#plt.xticks(range(0, 256*7+1,32*7)) 
# Optional: title
#fig.suptitle("NMOS vs. PMOS as switch (Simulated)")

ax1.grid(True, which='major', linestyle='--', linewidth=0.7, alpha=0.7)
ax1.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.5)

ax2.grid(True, which='major', linestyle='--', linewidth=0.7, alpha=0.7)
ax2.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.5)
plt.savefig(r"D:\ISSCC_26\Graphs\NMOS_PMOS_Switch.jpg", dpi=800, bbox_inches="tight")
plt.show()
```
