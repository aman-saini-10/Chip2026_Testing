```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset
from scipy.stats import norm
from matplotlib.ticker import MultipleLocator
```

```python
A0 = np.arange(0,8).astype(np.uint8)
A1 = np.arange(1,9).astype(np.uint8)
A2 = np.arange(2,10).astype(np.uint8)
A3 = np.arange(3,11).astype(np.uint8)
A4 = np.arange(4,12).astype(np.uint8)
A5 = np.arange(5,13).astype(np.uint8)
A6 = np.arange(6,14).astype(np.uint8)
A7 = np.arange(7,15).astype(np.uint8)
arrays = [A0, A1, A2, A3, A4, A5, A6, A7]
```

```python
plt.figure(figsize=(4,2))

for i, arr in enumerate(arrays):
    plt.plot(arr, [i]*len(arr), "o",color = "#1f77b4")  

plt.xlabel("$X_b$",fontsize = 12)
plt.ylabel("$W_b$",fontsize = 12)
plt.title("Hybrid Compute Model",fontsize = 12)
plt.ylim(0,8)
plt.xlim(0,16)
plt.xticks(np.arange(-1, 16, 1)) 
plt.yticks(np.arange(-1, 8, 1))
plt.grid(True)
plt.savefig("D:\ISSCC26\Dot_Matrix.jpg",dpi=800, bbox_inches="tight", transparent=True)
plt.show()
```
