```python

VB_VOLTAGES = [3,4,6,9]
VB_CALIB_VOLTAGES = np.load("D:\Chip2025\Chip2025_Testing\Python_Notebook\Tests\Data\OSC_Calibration\VB_CALIB_VOLTAGES.npy")
print(VB_CALIB_VOLTAGES.shape) #BANK0->BANK4 and Bottom to Top 4*72

BANK0_DEL_CALIB = np.zeros((72*4,32),dtype=np.uint8)
BANK1_DEL_CALIB = np.zeros((72*4,32),dtype=np.uint8)
BANK2_DEL_CALIB = np.zeros((72*4,32),dtype=np.uint8)
BANK3_DEL_CALIB = np.zeros((72*4,32),dtype=np.uint8)

for dl in range(72*4):
    for dc in range(32):
        if np.floor(dl/72) == 0: #BANK0
            ARR = BANK0_DEL_CALIB
            reversed = True
        if np.floor(dl/72) == 1: #BANK1
            ARR = BANK1_DEL_CALIB
            reversed = False
        if np.floor(dl/72) == 2: #BANK2
            ARR = BANK2_DEL_CALIB
            reversed = True
        if np.floor(dl/72) == 3: #BANK3
            ARR = BANK3_DEL_CALIB
            reversed = False
        k = dl%72
        vb = VB_CALIB_VOLTAGES[dl,dc]
        vb_ndx = VB_VOLTAGES.index(vb)
        if not reversed:
            ARR[(71-k)*4][dc] = 0 if (vb_ndx == 0 or vb_ndx == 2) else 1     #CAL_0
            ARR[(71-k)*4+1][dc] = 1 if (vb_ndx == 0 or vb_ndx == 2) else 0   #CAL_B_0
            ARR[(71-k)*4+2][dc] = 0 if (vb_ndx == 0 or vb_ndx == 1) else 1   #CAl_1
            ARR[(71-k)*4+3][dc] = 1 if (vb_ndx == 0 or vb_ndx == 1) else 0   #CAl_B_1
        if reversed:
            ARR[(71-k)*4][31-dc] = 0 if (vb_ndx == 0 or vb_ndx == 2) else 1     #CAL_0
            ARR[(71-k)*4+1][31-dc] = 1 if (vb_ndx == 0 or vb_ndx == 2) else 0   #CAL_B_0
            ARR[(71-k)*4+2][31-dc] = 0 if (vb_ndx == 0 or vb_ndx == 1) else 1   #CAl_1
            ARR[(71-k)*4+3][31-dc] = 1 if (vb_ndx == 0 or vb_ndx == 1) else 0   #CAl_B_1
```

```python
#Create the excel for SRAM WRITE
VBN_VOLTAGES = [5,6,7,8]
VBN_CALIB_VOLTAGES = np.load("D:\Chip2025\Chip2025_Testing\Python_Notebook\Tests\Data\OSC_Calibration\VBN_CALIB_VOLTAGES.npy")
VBP_VOLTAGES = [2,4,6,8]
VBP_CALIB_VOLTAGES = np.load("D:\Chip2025\Chip2025_Testing\Python_Notebook\Tests\Data\OSC_Calibration\VBP_CALIB_VOLTAGES.npy")
print(VBN_CALIB_VOLTAGES.shape) #Direction is BANK0,BANK1,BANK2,BANK3 within that from OSC 63 to 0. Bottom->Top
Refline_NDX = [4,13,22,31,40,49,58,67]
BANK0_OSC_CALIB = np.zeros(72*4,dtype=np.uint8)
BANK1_OSC_CALIB = np.zeros(72*4,dtype=np.uint8)
BANK2_OSC_CALIB = np.zeros(72*4,dtype=np.uint8)
BANK3_OSC_CALIB = np.zeros(72*4,dtype=np.uint8)
k = 0
m = 0
for dl in range(72*4):
    if np.floor(dl/72) == 0: #BANK0
        ARR = BANK0_OSC_CALIB
    if np.floor(dl/72) == 1: #BANK1
        ARR = BANK1_OSC_CALIB
    if np.floor(dl/72) == 2: #BANK2
        ARR = BANK2_OSC_CALIB
    if np.floor(dl/72) == 3: #BANK3
        ARR = BANK3_OSC_CALIB
    k = dl%72
    if k not in Refline_NDX:
        vbn = VBN_CALIB_VOLTAGES[m]
        vbn_ndx = VBN_VOLTAGES.index(vbn)
        vbp = VBP_CALIB_VOLTAGES[m]
        vbp_ndx = VBP_VOLTAGES.index(vbp)
        ARR[(71 - k)*4] = 0 if (vbn_ndx == 0 or vbn_ndx == 2) else 1    #NCAL0
        ARR[(71 - k)*4+1] = 0 if (vbn_ndx == 0 or vbn_ndx == 1) else 1  #NCAL1
        ARR[(71 - k)*4+2] = 0 if (vbp_ndx == 0 or vbp_ndx == 2) else 1  #PCAL0
        ARR[(71 - k)*4+3] = 0 if (vbp_ndx == 0 or vbp_ndx == 1) else 1  #PCAL1
        m = m+1
```
