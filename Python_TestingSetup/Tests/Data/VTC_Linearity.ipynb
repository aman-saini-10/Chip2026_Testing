```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.axes_grid1.inset_locator import inset_axes, mark_inset
from scipy.interpolate import interp1d
from matplotlib.ticker import MultipleLocator
```

```python
import pandas as pd
df = pd.read_csv(r"d:\Chip2025_Testing\Python_Notebook\Tests\Data\CadenceGraphs\VTC_LIN.csv")

data_TT = df["C0"][6::2].to_numpy(dtype = "float")
data_TT = data_TT.reshape(11,31)
data_FF = df["C1"][6::2].to_numpy(dtype = "float")
data_FF = data_FF.reshape(11,31)
data_SS = df["C2"][6::2].to_numpy(dtype = "float")
data_SS = data_SS.reshape(11,31)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.rcParams['font.family'] = 'Times New Roman'   # or 'Times New Roman', 'DejaVu Sans', etc.
mpl.rcParams['font.size'] = 12

# Axis-specific defaults
mpl.rcParams['axes.titlesize'] = 14
mpl.rcParams['axes.labelsize'] = 14
mpl.rcParams['xtick.labelsize'] = 12
mpl.rcParams['ytick.labelsize'] = 12
mpl.rcParams['legend.fontsize'] = 10
```

```python
fig, ax1 = plt.subplots(1, 1, figsize=(4, 2),sharex=True)

x_axis = np.arange(600,910,10)
for i in range(0,len(data_TT)):
    ax1.plot(x_axis,(data_TT[i] - data_TT[i][0])*2*10**9, color="g",linewidth=1, label = "", alpha = 0.3)
    ax1.plot(x_axis,(data_FF[i] - data_FF[i][0])*2*10**9, color="b",linewidth=1, label = "", alpha = 0.3)
    ax1.plot(x_axis,(data_SS[i] - data_SS[i][0])*2*10**9, color="r",linewidth=1, label = "", alpha = 0.3)   

ax1.plot(x_axis,(data_TT[5] - data_TT[5][0])*2*10**9, color="g",linewidth=2, alpha = 1, label = "TT  27° 0.9V")
ax1.plot(x_axis,(data_FF[5] - data_FF[5][0])*2*10**9, color="b",linewidth=2, alpha = 1, label = "FF   0°   1V")
ax1.plot(x_axis,(data_SS[2] - data_SS[2][0])*2*10**9, color="r",linewidth=2, alpha = 1, label = "SS 100° 0.8V")

ax1.legend(framealpha = 0.9)
ax1.set_ylabel("Delay (ns)")
ax1.set_xlabel("VCBL (mV)")
#fig.suptitle("VTC delay correction across PVT corners (Simulated)")
ax1.grid(True, which='major', linestyle='--', linewidth=0.7, alpha=0.7)
ax1.grid(True, which='minor', linestyle=':', linewidth=0.5, alpha=0.5)
plt.savefig(r"D:\ISSCC_26\Graphs\VTC_LIN.jpg", dpi=800, bbox_inches="tight")
```
