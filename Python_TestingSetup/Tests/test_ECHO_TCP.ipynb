```python
import numpy as np
import sys
sys.path.append(r"../Communiaction")
import TCP_IP_UTILS
from collections import Counter
```

```python
HOST = '192.168.1.10'  # Remote device IP (server IP)
PORT = 7               # Echo port
```

```python
def print_element_counts(lst):
    counts = Counter(lst)
    for value, count in counts.items():
        print(f"Value {value} occurs {count} times")

def compare_lists(list1, list2):
    if len(list1) != len(list2):
        print("Failure: Lists are of different lengths.")
        print(len(list1))
        return
    fail = 0
    for i, (a, b) in enumerate(zip(list1, list2)):
        if a != b:
            print(f"Failure: Mismatch at index {i}: {a} != {b}")
            fail = 1
            
    if(fail):
        return 1
    print("Success: All elements match.")
    return 0
```

```python
def echo(client_socket):
        total_length = 5000000
        random_array = np.random.randint(0, 256, size=total_length, dtype=np.uint8)
        data_to_send = bytearray(random_array)
        #call the function along with data
        data = TCP_IP_UTILS.call_fpga_fn(client_socket,0,total_length,data_to_send,0)
        ret = compare_lists(data,list(data_to_send))
        return ret
```

```python
check_list = [];
client_socket = TCP_IP_UTILS.tcp_connect(HOST,PORT)
for i in range(0,10):
    check_list.append(echo(client_socket))
print_element_counts(check_list)
```
